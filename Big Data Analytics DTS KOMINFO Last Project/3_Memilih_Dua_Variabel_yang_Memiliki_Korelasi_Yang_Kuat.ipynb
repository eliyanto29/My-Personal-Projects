## Memilih Dua Atribut Yang Memiliki Korelasi Besar

### Problem Statement:
Memilih 2 atribut yang diduga memiliki korelasi yang kuat.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
databersih=pd.read_csv('databersih.csv')
```

```python
plt.figure(figsize=(12,8))
#Ploting
plt.title('Kompensasi Karyawan Kota San Fransisco (USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(databersih.index, databersih.salaries, 'b.-', label='Gaji Pokok')
plt.plot(databersih.index, databersih.overtime, 'g.-', label='Gaji Lembur')
plt.plot(databersih.index, databersih['other salaries'], 'r.-', label='Gaji Tambahan')
plt.plot(databersih.index, databersih.retirement, 'm.-', label='Pensiunan')
plt.plot(databersih.index, databersih['health and dental'], 'c.-', label='Tunjangan Kesehatan')
plt.plot(databersih.index, databersih['other benefits'], 'k.-', label='Tunjangan Lainnya')

plt.xlabel('Tahun')
plt.ylabel('Dolar US')

plt.legend()
plt.savefig('Kompensasi Karyawan Kota San Fransisco.png', dpi=300)

plt.show()
```

#### Catatan:
Berdasarkan Grafik di atas maka dapat dilihat bentuk yang mirip antara 6 variabel yaitu gaji pokok, gaji lembur, gaji tambahan, pensiunan, tunjangan kesehatan dan tunjangan lainnya.

```python
import numpy as np
import pandas as pd
datanumber=databersih[['salaries','overtime','retirement','other salaries', 'total salary', 'health and dental','other benefits', 'total benefits', 'total compensation']]
Korelasi=datanumber.corr()
print(Korelasi)
```

### Catatan:
Berdasarkan hasil penghitungan nilai korelasi di atas, dapat dilihat bahwa terdapat korelasi yang sangat tinggi(>0,9) antara:
1. Total Salary(Gaji Keseluruhan) dan Total Compensation(Kompensasi Keseluruhan) sebesar 0.991156
2. Salaries(Gaji Pokok) dan Total_Compensation(Kompensasi Keseluruhan) sebesar 0.968729
2. Salaries(Gaji Pokok) dan Total_Salary(Gaji Keseluruhan) sebesar 0. 0.960858
3. Total Compensation(Kompensasi Keseluruhan) dan Retirement(Pensiunan) sebesar 0.952875 
4. Salaries(Gaji Pokok) dan Retirement(Pensiunan) sebesar 0.948921 
5. Total Compensation(Kompensasi Keseluruhan) dan Total Benefits(Total Tunjangan) sebesar  0.930141
6. Total Benefits(Total Tunjangan) dan Retirement(Pensiunan) sebesar 0.923618

## Scatter Plot antara Gaji Keseluruhan dan Kompensasi Keseluruhan

```python
import matplotlib.pyplot as plt
import numpy as np

#Menentukan ukuran gambar
plt.figure(figsize=(12,8))
#Ploting
plt.title('Gaji VS Kompensasi', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.scatter(databersih['total salary'], databersih['total compensation'])

plt.xlabel('Gaji')
plt.ylabel('Kompensasi')

plt.savefig('Gaji VS Kompensasi.png', dpi=300)

plt.show()
```

### Catatan:
Berdasarkan hasil plot di atas nampak bahwa Total Kompensasi sangat dipengaruhi oleh Total Gaji, dengan derajat korelasi yaitu 0.991156. Hal itu dapat diterima, karena semakin besar gaji seorang karyawan maka semakin besar kompensasi yang ia terima.
