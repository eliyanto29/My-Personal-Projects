# Tugas Akhir BDA DTS 

Data kompensasi karyawan Kota San Fransisco merupakan data yang dikumpulkan oleh pihak yang berwenang di kota San Fransisco tentang besaran Kompensasi karyawan dengan berbagai jenis pekerjaan. Data ini diunggah secara resmi di http://openbook.sfgov.org. Data ini diambil mulai tahun 2013, data diperbaharui secara berkala. 

Link sumber data: https://www.kaggle.com/san-francisco/sf-employee-compensation#employee-compensation.csv

## Keterangan Detail Atribut Data

**Jenis Tahun:**
Fiskal (Juli hingga Juni) atau Kalender (Januari hingga Desember)

**Tahun:**
Periode akuntansi 12 bulan. Kota San Francisco beroperasi pada tahun fiskal yang dimulai pada 1 Juli dan berakhir pada 30 Juni tahun berikutnya. Tahun Anggaran yang berakhir 30 Juni 2012 diwakili sebagai Tahun Fiskal 2011-2012.

**Kode Grup Organisasi:**
Grup Organisasi adalah sekelompok Departemen. Misalnya, Grup Organisasi Perlindungan Publik mencakup departemen seperti Kepolisian, Pemadam Kebakaran, Jaksa Distrik, dan Sheriff.

**Kelompok Organisasi:**
Grup Organisasi adalah sekelompok Departemen. Misalnya, Grup Organisasi Perlindungan Publik mencakup departemen seperti Kepolisian, Pemadam Kebakaran, Jaksa Distrik, dan Sheriff.

**Kode Departemen:**
Departemen adalah unit organisasi utama yang digunakan oleh Kota San Francisco. Contohnya Departemen Rekreasi dan Taman, Pekerjaan Umum, dan Departemen Kepolisian.

**Departemen:**
Departemen adalah unit organisasi utama yang digunakan oleh Kota dan Kabupaten San Francisco. Departemen termasuk Rekreasi dan Taman, Pekerjaan Umum, dan Departemen Kepolisian.

**Kode Serikat Pekerja:**
Serikat pekerja mewakili karyawan dalam perjanjian perundingan bersama. suatu pekerjaan milik satu serikat pekerja, meskipun beberapa pekerjaan tidak terwakili (biasanya sementara).

**Serikat Pekerja:**
Serikat pekerja mewakili karyawan dalam perjanjian perundingan bersama. Suatu pekerjaan milik satu serikat pekerja, meskipun beberapa pekerjaan tidak terwakili (biasanya sementara).

**Kode Kelompok Pekerjaan:**
Pekerjaan yang serumpun menggabungkan beberapa pekerjaan yang serupa ke dalam kelompok pekerjaan.

**Kelompok Pekerjaan:**
Pekerjaan yang serumpun menggabungkan beberapa pekerjaan yang serupa ke dalam kelompok pekerjaan.

**Kode Pekerjaan:**
Pekerjaan ditentukan oleh unit klasifikasi Sumber Daya Manusia. Contohnya: tukang kebun, petugas polisi, dan akuntan.

**Pekerjaan:**
Pekerjaan ditentukan oleh unit klasifikasi Sumber Daya Manusia. Contohnya: tukang kebun, petugas polisi, dan akuntan.

**Pengidentifikasi Karyawan:**
Setiap angka berbeda di kolom “Identifier Pegawai” mewakili satu karyawan. Angka-angka pengidentifikasi ini tidak bermakna tetapi secara acak ditugaskan untuk tujuan membangun dataset ini. Kolom tidak muncul pada laporan Kompensasi Karyawan yang di-host di openbook.sfgov.org, tetapi laporan itu menunjukkan satu baris untuk setiap karyawan. ID Karyawan telah disertakan di sini untuk memungkinkan pengguna merekonstruksi laporan asli. Perhatikan bahwa pengidentifikasi setiap karyawan akan berubah setiap kali dataset ini diperbarui, jadi perbandingan oleh karyawan di beberapa versi dataset tidak dimungkinkan.

**Gaji:**
Gaji normal dibayarkan kepada karyawan secara permanen atau sementara.

**Lembur:**
Jumlah yang dibayarkan kepada karyawan yang bekerja lebih dari 40 jam per minggu.

**Gaji lainnya:**
Berbagai pembayaran tidak teratur yang dilakukan kepada karyawan seperti pembayaran premi, pembayaran insentif, atau pembayaran satu kali lainnya.

**Total Gaji:**
Jumlah semua gaji dibayarkan kepada karyawan.

**Pensiunan:**
Adalah sejumlah tunjangan untuk karyawan yang sudak memasuki masa pensiun dari pemerintah kota.

**Tunjangan Kesehatan dan Gigi:**
Premi yang dibayarkan ke kota untuk paket asuransi kesehatan dan gigi yang mencakup karyawan Kota. Untuk melindungi kerahasiaan sebagaimana disyaratkan secara hukum, rata-rata seluruh kota yang diberi peringkat diperlihatkan sebagai pengganti tunjangan kesehatan dan gigi khusus karyawan.

**Tunjangan lain:**
Tunjangan wajib dibayarkan atas nama karyawan, seperti kontribusi Jaminan Sosial (FICA dan Medicare).

**Total Tunjangan:**
Jumlah dari semua tunjangan karyawan.

**Total Kompensasi:**
Jumlah semua bentuk kompensasi, baik gaji maupun tunjangan.

## Data *Cleaning*

```python
import pandas as pd
data=pd.read_csv('employee-compensation.csv')
data.head()
```

```python
#Memilih data gaji pokok yang bukan nol
datap=data[data.Salaries != 0]
#Membuat data yang negatif menjadi positif
salaries={}
for baris in datap['Salaries']:
    if baris < 0:
        salaries=abs(datap['Salaries'])

overtime={}
for baris in datap['Overtime']:
    if baris < 0:
        overtime=abs(datap['Overtime'])

other_salaries={}
for baris in datap['Other Salaries']:
    if baris < 0:
        other_salaries=abs(datap['Other Salaries'])

total_salary={}
for baris in datap['Total Salary']:
    if baris < 0:
        total_salary=abs(datap['Total Salary'])

retirement={}
for baris in datap['Retirement']:
    if baris < 0:
        retirement=abs(datap['Retirement'])

health_and_dental={}
for baris in datap['Health and Dental']:
    if baris < 0:
        health_and_dental=abs(datap['Health and Dental'])

other_benefits={}
for baris in datap['Other Benefits']:
    if baris < 0:
        other_benefits=abs(datap['Other Benefits'])

total_benefits={}
for baris in datap['Total Benefits']:
    if baris < 0:
        total_benefits=abs(datap['Total Benefits'])

total_compensation={}
for baris in datap['Total Compensation']:
    if baris < 0:
        total_compensation=abs(datap['Total Compensation'])
```

```python
#Membentuk DataFrame Baru
databersih=pd.DataFrame({'salaries': salaries,'overtime': overtime,'retirement': retirement,'other salaries':other_salaries, 'total salary':total_salary, 'health and dental':health_and_dental, 'other benefits':other_benefits,'total benefits':total_benefits,'total compensation':total_compensation})
databersih.head()
```

```python
data.head()
```

```python
#Mengganti Data Number Salaries sampai Total Compensation dengan data yang sudah dibersihkan
databaru=datap.drop('Salaries',axis=1)
databaru=databaru.drop('Overtime',axis=1)
databaru=databaru.drop('Other Salaries',axis=1)
databaru=databaru.drop('Total Salary',axis=1)
databaru=databaru.drop('Retirement',axis=1)
databaru=databaru.drop('Health and Dental',axis=1)
databaru=databaru.drop('Other Benefits',axis=1)
databaru=databaru.drop('Total Benefits',axis=1)
databaru=databaru.drop('Total Compensation',axis=1)
dataclean=pd.concat([databaru, databersih], axis=1)
dataclean.head()
```

```python
# Menghapus data yang terduplikasi
databersih=dataclean.drop_duplicates(keep=False)
```

```python
# mengkonversi dataframe ke file csv
bytes_to_write = databersih.to_csv('databersih.csv')
```
