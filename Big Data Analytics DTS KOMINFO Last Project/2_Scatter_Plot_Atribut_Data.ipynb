## Scater Plot Atribut Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
databersih=pd.read_csv('databersih.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Menentukan ukuran gambar
plt.figure(figsize=(12,8))
#Ploting
plt.title('Gaji Karyawan di Kota San Fransisco', fontdict={'fontweight':'bold', 'fontsize': 18})
#float64(data['Salaries'])
plt.scatter(databersih.index, databersih['salaries'])

plt.xlabel('Pekerjaan')
plt.ylabel('Gaji')

plt.savefig('Gaji Karyawan di Kota San Fransico.png', dpi=300)

plt.show()
```
